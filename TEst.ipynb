```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv('MetroPT3(AirCompressor).csv')
```

```python
data.head(4)
```

```python
data.describe().T
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
data.head(4)
```

```python
data = data.drop(['COMP', 'DV_eletric', 'Towers', 'MPG', 'LPS', 'Pressure_switch', 'Oil_level', 'Caudal_impulses'], axis = 1)
```

```python
data.head(4)
```

```python
data_to_normalize = data.drop('timestamp', axis=1)

# Создание объекта нормализатора
scaler = MinMaxScaler()

# Применение нормализации к данным
normalized_data = scaler.fit_transform(data_to_normalize)

# Создание нового датафрейма с нормализованными данными
normalized_data = pd.DataFrame(normalized_data, columns=data_to_normalize.columns)

# Объединение столбца 'timeshtamp' и нормализованных данных
final_data = pd.concat([data['timestamp'], normalized_data], axis=1)

# Вывод окончательного датафрейма
final_data.head(4)
```

```python
data_for_clustering = final_data.drop('timestamp', axis=1)

# Пустые списки для сохранения значений инерции и коэффициента силуэта
inertias = []
silhouette_scores = []

# Попробовать различные значения количества кластеров от 2 до 10
for n_clusters in range(2, 11):
    # Применение алгоритма K-means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_for_clustering)
    
    # Сохранение инерции и коэффициента силуэта
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(data_for_clustering, kmeans.labels_))

# Визуализация метода локтя
plt.plot(range(2, 11), inertias, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.title('Метод локтя')
plt.show()

# Визуализация метода силуэта
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Коэффициент силуэта')
plt.title('Метод силуэта')
plt.show()
```
