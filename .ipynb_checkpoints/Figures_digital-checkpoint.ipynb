```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('MetroPT3(AirCompressor).csv')
```

```python
data.head(4)
```

```python
data.describe().T
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
start_date = pd.to_datetime('4/18/2020', format='%m/%d/%Y')
end_date = start_date + pd.DateOffset(days=1)
filtered_data = data[(data['timestamp'] >= start_date) & (data['timestamp'] < end_date)]

plt.plot(filtered_data['timestamp'], filtered_data['COMP'])

plt.xlabel('Временной штамп')
plt.ylabel('Значение COMP')
plt.title('График значения столбца COMP для 4/18/2020')

plt.show()
```
